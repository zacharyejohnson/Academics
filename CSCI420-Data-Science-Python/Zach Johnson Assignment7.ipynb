```python
''' 
Add codes to this cell.
Use the "company_sale_data.csv" as input
Generate a line plot showing total profit by month
The generated line plot must include the following properties:

X label name = Month Number
Y label name = Total profit
Figure title=Company profit per month

'''
import matplotlib.pyplot as plt
import pandas as pd 

data = pd.read_csv("company_sales_data.csv")
data.plot(kind = "line", y = "total_profit", x="month_number", xlabel="Month Number",
          ylabel = "Total Profit", title="Company Profit Per Month", legend=False, figsize=(10,5))

```

```python
''' 
On a new figure, draw a line plot showing Total Units by month.
The plot must include the following properties: 

Line Style should be dotted and Line-color should be red
Show legend at the lower right location, Legend=Profit data of last year

X label name = Month Number
Y label name = Total Units
Add a circle marker.
Line marker color as red
Line width should be 3
Sales data=Units sold per month
'''
data.plot(x ="month_number", y="total_units", style="o--r", lw =3, xlabel="Month Number",
          ylabel="Total Units", title="Units Sold Last Month", figsize = (10,5))
plt.legend(["Profit Data of Last Year"], loc='lower right')
```

```python
'''
The CSV file has six products, which are 
facecream, facewash,toothpaste,bathingsoap,shampoo,moisturizer

On a new figure,
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product (column) ).
Each line

Line Style should be solid,
Add a circle marker.
Show legend at the upper location, Use corresponding column names as line legends

X label name = Month Number
Y label name = Units Sold
Add a circle marker
Line width should be 3
Figure title=Sales data
'''
data.plot(x="month_number", y=["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"],
          style="-o", lw=3, title="Sales Data", ylabel="Units Sold", xlabel="Month Number", figsize = (10,5))
plt.legend(loc="upper left")
```

```python
'''
On a new figure, Calculate total units sold for last year for each product and show it using a Pie chart
The Pie chart displays the percentages of total units that each product accounts for
E.g. 
If a total of 100 units was sold last year (counting all products), and a total of 10 uints of product A was sold in the last year. 
Then the pie chart should show that product A account for 10.0%, round up to one digit after the decimal point

For documentation of pie chart
https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

'''

#playing around with the shadow a bit- I hope you like it! Thanks for a great semester. 

plt.rcParams.update({'font.size': 12})
data[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]].sum().plot(kind='pie', shadow = True, startangle=90,
figsize=(15,10), autopct='%1.1f%%', labels = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"], ylabel=" ")
```
