```python
import numpy as np
import pandas as pd
import pandas_datareader as web
import datetime as dt
import statsmodels.api as sm
import seaborn as sb
import matplotlib.pyplot as plt
import statsmodels.stats.diagnostic as smd
import statsmodels.tsa.stattools as stools
import statsmodels.graphics.gofplots as qqplot
```

```python
data_codes = {"IP": "INDPRO",
              "EI": "EXPINF1YR", 
              "I": "CORESTICKM159SFRBATL",
              "UPR": "BAA10Y",
              "LGB": "IRLTLT01USM156N",
              "TB": "DGS10", 
              "PCE": "PCE", 
              "OP": "DCOILWTICO"}
```

```python
start = dt.datetime(1980, 1, 1)
end = dt.datetime(2022, 12, 1)
data = pd.DataFrame(columns = data_codes.keys())
for key, val in data_codes.items(): 
    dataset = web.DataReader(val, 'fred', start, end)[val]
    data[key] = dataset
```

```python
data["UI"] = data["I"] - data["EI"].shift(-12)
```

```python
data["UTS"] = data["LGB"] - data["TB"].shift(-1)
```

```python
sp500 = web.DataReader("^GSPC", "yahoo", start, end)
data["SP500"] = sp500["Adj Close"]
```

```python
data= data.dropna()
for key in data.keys(): 
    data["d"+key] = data[key].pct_change()
```

```python
fig, ax = plt.subplots(figsize = (7,5))
sb.histplot(data["dPCE"])
```

```python
exog_vars = ["dIP", "dEI", "dUPR", "UI","UTS", "dOP", "dSP500"]
reg_data = data[exog_vars].dropna()
reg_data.corr().to_excel("data_description.xlsx", sheet_name="datacorr")

```

```python
lagged_data = data[exog_vars]
for var in exog_vars:
    lagged_data[var+'lag1'] = 0
    lagged_data[var+'lag2'] = 0
    lagged_data[var+'lag3'] = 0
    lagged_data[var+'lag4'] = 0
    lagged_data[var+'lag1'] = lagged_data.shift(-1)
    lagged_data[var+'lag2'] = lagged_data.shift(-2)
    lagged_data[var+'lag3'] = lagged_data.shift(-3)
    lagged_data[var+'lag4'] = lagged_data.shift(-4)
```

```python
stock_data = {sector: {} for sector in sectors}
for sector in sectors: 
    stock_data[sector] = pd.read_excel("SP_stock_data.xlsx", sheet_name = sector, parse_dates=True, index_col=0)
```

```python
sector_changes = {sector: pd.DataFrame() for sector in sectors}
for sector in sectors:
    for stock in stock_data[sector]:
        data2 = stock_data[sector][stock].reindex_like(data).pct_change()
        sector_changes[sector][stock] = data2
    
```

```python
avg_sector_changes = {df: sector_changes[df].mean(axis=1) for df in sector_changes.keys()}
```

```python
avg_sector_changes
```

```python
results_dict = {}
SC_test_dict = {}
f_model_dict = {}
param_vals = {sector: {b: [] for b in exog_vars} for sector in sectors}
for sector in sectors:
    cleaning_vars = ["dIP", "dEI", "dUPR", "UI","UTS", "dOP", "dSP500", sector]
    data[sector] = avg_sector_changes[sector]
    reg_data = data[cleaning_vars].dropna()
    X = reg_data[exog_vars]
    Y = reg_data[sector]
    X = sm.add_constant(X)
    model = sm.OLS(Y, X).fit()
    print(model.summary())
    fX = reg_data[["dIP", "dEI", "dUPR", "UI","UTS", "dOP"]]
    fY = reg_data[sector]
    fmodel = sm.OLS(fY, fX).fit()
    f_model_dict[sector] = fmodel
    results_dict[sector] = model
    reg_data["u"] = model.resid
    reg_data["u(t-1)"] = reg_data["u"].shift(-1).fillna(0)
    X1 = reg_data[["dIP", "dEI", "dUPR", "UI","UTS", "dOP", "dSP500", "u(t-1)"]].dropna()
    Y1 = reg_data["u"]
    X1 = sm.add_constant(X1)
    SCtest_model = sm.OLS(Y1, X1).fit()
    SC_test_dict[sector] = SCtest_model
    
        
plt.legend=True
```
