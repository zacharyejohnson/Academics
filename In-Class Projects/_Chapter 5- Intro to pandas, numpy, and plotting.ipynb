### Lists can have multiple dimensions 

```python
# can access multidimensional list using
# square brackets
a = [ [2, 4, 6, 8 ],     # lists within lists- kind of like dictionaries 
    [ 1, 3, 5, 7 ],
    [ 8, 6, 4, 2 ],
    [ 7, 5, 3, 1 ] ]

for i in range(len(a)) :
    for j in range(len(a[i])) :
        print(a[i][j], end=" ")
    print()	

```

#### these multi-dimesion lists can be updated using their indeces 

```python
import numpy as np

a = np.zeros((4,4)) # this creates 4x4 matrix with all values = 0
print("Origional Array: ", a, sep = "\n")

a[0][0] = 16     #specify index to be updated 
a[0][1] = 10
a[0][2] = 5
a[0][3] = 9
a[1][1] = 3
a[2][1] = 7

print("Updated Array: ", a, sep = "\n")

# super updated array 
print("\n Super updated array with nested for loops : ")
for i in range(len(a)):
    for j in range(len(a[i])): 
        a[i][j] = (i + 4)*5    ## random change 
print(a)
    
```

### there are several numpy commands that will instantiate and fill arraysd of various values 

```python
import numpy as np

lists = [[2,4,6],[1,3,5],[5,10,15]]  # create new list
arrays = np.array(lists)        # change lists into a np array
zeros = np.zeros_like(lists)     # create array of size of lists but fill with zeros 
ones = np.ones_like(lists)      # create array of size of lists but fill with ones 
empty = np.empty_like(lists)    # create empty array 

print("lists \n", lists)
print("arrays \n", arrays)
print("zeros \n", zeros)
print("ones \n", ones)
print("empty \n", empty)
```

# MONTY CARLO SIMULATION

```python
# montecarlo simulation
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt  
import random  


sim_length = 1000
random_list = []
for i in range (sim_length):
    random_list.append(random.normalvariate(0, .5))
    
plt.plot(random_list, ls = "", marker = ".", markersize = 1)
plt.title('Monte Carlo Simulation')
plt.show()
plt.close()
                       
                       
                       
                       


                       
                       
```

```python
simulation = []

for i in range(len(random_list)): 
    val = random_list[i]
    if i == 0: 
        simulation.append(val)
    else: 
        simulation.append(simulation[i-1]+val)
        
        
        
plt.plot(simulation)

plt.show()
plt.close()
```

```python
monte_carlo_sim_dict = {}
num_sims = 200
periods = 100
mean = 0
sigma = 1
for i in range(num_sims): 
    monte_carlo_sim_dict[i] = {}
    for period in range(periods): 
        curr_sim = monte_carlo_sim_dict[i]
        if period == 0: 
            curr_sim[period] = random.normalvariate(mean, sigma)
        else: 
            curr_sim[period] = curr_sim[period-1] + random.normalvariate(mean, sigma)
monte_carlo_sim_dict
monte_carlo_sim_df =  pd.DataFrame(monte_carlo_sim_dict)
```

```python
monte_carlo_sim_df.plot.line(legend = False)
```
