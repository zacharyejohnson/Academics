1.	Calculate the real base money stock (M/P) using data for the GDP Deflator and the Monetary Base. Make sure to reindex the price level by dividing it by 100.

$(1/1/100) --> 1 * 100$

```python
import pandas_datareader.data as web
import datetime as dt
from datlib.FRED import *

start = dt.datetime(1975, 1, 1)
data_codes = {"GDP Deflator": "GDPDEF", "Monetary Base": "BOGMBASEW"}
data = gather_data(data_codes, start = start)
GDP_deflator = data["GDP Deflator"].div(100)
data["Real Monetary Base"] = data["Monetary Base"].div(GDP_deflator)
data
```

2.	Log the values of the Monetary Base and the Real Monetary Base

```python
import numpy as np
log_data = np.log(data)
data["Logged GDP Deflator"] = np.log(data["GDP Deflator"])
data["Logged Monetary Base"] = np.log(data["Monetary Base"])
data["Logged Real Monetary Base"] = np.log(data["Real Monetary Base"])
data
```

3.	Calculate the First Difference (log(xt) â€“ log(x t-1) of Logged Values for the Monetary Base and Real Monetary Base

```python
# hint: use log_data.diff()
data["Monetary Base Logged Difference"] = data["Logged Monetary Base"].diff(1)
data["Real Monetary Base Logged Difference"] = data["Logged Real Monetary Base"].diff(1)
data = data.drop(columns=["Logged Real Monetary Base Logged Difference", "Logged Monetary Base Logged Difference"] )
data
```
