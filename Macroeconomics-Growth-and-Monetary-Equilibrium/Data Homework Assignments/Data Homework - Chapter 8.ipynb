1.	Generate logged values for each variable except for the 10 Year U.S. Treasury Rate.

```python
from datlib.FRED import *
import datetime

start = datetime.datetime(1970,1,1)
data_codes = {
    "Base Money": "BOGMBASE",
    "Currency in Circulation": "WCURCIR",
    "M1": "M1",
    "M2": "M2",
    "Gross Domestic Private Investment": "GPDI",
    "Nominal GDP": "GDP",
    "10 Year U.S. Treasury Rate": "DGS10"
}

data = gather_data(data_codes, start = start)
```

8.	Is Nominal GDP correlated with GDPI? Why or why not?

```python
import statsmodels.api as sm

vars1 = ["Gross Domestic Private Investment Logged difference", "Nominal GDP Logged difference"]
results = sm.OLS(data[vars1[0]].dropna(), data[vars1[1]].dropna()).fit()
results.summary()
```

GDPI and Nominal GDP are highly correlated with an r2 of .615 and p value of ~0.000. As nominal GDP(amount of money in peoples pockets) increases, they are expected to invest more. This makes sense. More money means more investment. And from the equation of exchange, Nominal income is directly related to level of total expendiatures, which is the total amount of money multiplied by how fast it is moving.
